```python
import yfinance as yf
import pandas as pd
import sqlite3
```

```python
raw_data=yf.download("SPY",start="2015-01-01",end="2025-01-01")   #pulling data of SPDR S&P 500 ETF from yahoo finance
```

```python
raw_data.columns=raw_data.columns.get_level_values(0)   #remove the SPY subheader
raw_data.reset_index(inplace=True)   #move date from index to a seperate column
raw_data.head()
```

```python
conn=sqlite3.connect("risk_data.db")
raw_data.to_sql("stock_prices",conn,if_exists="replace",index=False)
print("Data connection successfull and saved to SQL")
```
